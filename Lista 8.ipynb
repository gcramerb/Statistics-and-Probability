```python
# Questão 9:
from sklearn import datasets
iris = datasets.load_iris()
df = pd.DataFrame(iris['data'],columns = ['sepal length','sepal width ','petal length','petal width'])
#iris = pd.DataFrame(iris)
sns.pairplot(df)
```

```python
setosa = pd.DataFrame(iris.data[iris.target==0])
setosa.columns= [ feature[:-4] for feature in iris.feature_names]

mean = setosa.mean()
print(f"Média \n{mean}\n")
cor = setosa.corr().round(2)
print(f'Correlação \n{cor}\n')
cov = setosa.cov().round(2)
print(f'Covariância \n{cov}\n')

print("Maior correlação é entre sepal length e sepal width, com 0,74")
print("Menor correlação é entre petal length e sepal width, com 0,18\n")

N2 = [mean[0:2], cor.iloc[:2,:2]]
print(N2)
```
